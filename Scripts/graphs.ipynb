Plot Port call statistics

```python
#create a line graph using port call statistics 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
#read the data from the csv
file= pd.read_csv("your-file-path")
df=pd.DataFrame(file,columns=['month','arrivals','departures'])
#print(df)

#plot the arrivals/departures
df = df.set_index('month')
plt.figure().set_figwidth(15)
col = ['arrivals', 'departures']
colors = ['#2B2F42', '#8D99AE']


for i,col in enumerate(col):
    plt.plot(df.index, df[col],marker='o',color=colors[i],linewidth=2,linestyle='solid')

plt.xlabel('Date')
plt.ylabel('Number of Vessels')
plt.title('Arrivals and Departures per month')
plt.legend(['Arrivals','Departures']) 
plt.show()
```

Interactive Plot

```python
#create an interactive plot
import plotly.express as px

df=pd.DataFrame(file,columns=['month','arrivals','departures'])

fig = px.line(df, x='month', y=df.columns[1:3],markers =True,hover_data=['month'],title='Arrivals and Departures per month',color_discrete_sequence=['#778AAE','#0D2A63'])

fig.update_layout(template='plotly_white',xaxis_title='Date',yaxis_title='Number of Vessels',legend_title='')

fig.show()
```
